```python
# So the request has succeeded

# We can use beautiful soup to parse structured data in HTML format
# Note that this would work for static web content
# If you want to scrape dynamic web content, you would have to use Selenium
# If you use requests and bs4 on static web content, you won't get a error, but a JavaScript code

from bs4 import BeautifulSoup

URL = 'https://www.monster.com/jobs/search/?q=software-developer&where=Australia'
page = requests.get(URL)

# create a BeautifulSoup object
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
results = soup.find(id='ResultsContainer')
print(results)
```

```python
# For easier viewing

print(results.prettify())
```

```python
# HTML/XML documents have a tree structure, dictated by DOM (Document Object Model)

# Find elements by using HTML class name

# Every job posting is wrapped in a <section> element with the class card-content

job_elems = results.find_all('section', class_='card-content')

for jobs in job_elems:
    print(jobs, end='\n'*2)
```

```python
# There is still a lot of clutter w.r.t the information that we need

for job_elem in job_elems:
    title_elem = job_elem.find('h2', class_='title')
    company_elem = job_elem.find('div', class_='company')
    location_elem = job_elem.find('div', class_='location')
    print(title_elem)
    print(company_elem)
    print(location_elem)
    print()
    
```

```python
# using .text with a bs4 object will return only the text part

for job_elem in job_elems:
    title_elem = job_elem.find('h2', class_='title')
    company_elem = job_elem.find('div', class_='company')
    location_elem = job_elem.find('div', class_='location')
    print(title_elem.text)
    print(company_elem.text)
    print(location_elem.text)
    print()
```

```python
# We still have whitespaces here, you will probably get an error 'NoneType' object has no attribute 'text'
# The structure of the web is really messy, so there might be other elements too that might have crept in
# Let's remove the undesireable elements (This could be ads or other extraneous stuff)
# use strip method in python to remove the whitespace

for job_elem in job_elems:
    title_elem = job_elem.find('h2', class_='title')
    company_elem = job_elem.find('div', class_='company')
    location_elem = job_elem.find('div', class_='location')
    if None in (title_elem, company_elem, location_elem):
        continue
    print(title_elem.text.strip())
    print(company_elem.text.strip())
    print(location_elem.text.strip())
    print()
```

```python
# Now we have a really good, structured, easily readable data

# Let's toy around a bit

from collections import defaultdict

company_info = {}



for job_elem in job_elems:
    title_elem = job_elem.find('h2', class_='title')
    company_elem = job_elem.find('div', class_='company')
    location_elem = job_elem.find('div', class_='location')
    if None in (title_elem, company_elem, location_elem):
        continue
    title_elem = title_elem.text.strip()
    company_elem = company_elem.text.strip()
    location_elem = location_elem.text.strip()
    company_info[company_elem] = (title_elem, location_elem)
    
```

```python
for companies in company_info:
    print(companies, company_info[companies])
    print('\n')
```

```python
# We can write functions/method to make this more moular and efficient, reusable to use

# Find number of jobs in Sydney

sydney_jobs = 0
key = "Sydney"

for companies in company_info:
    # print(companies, company_info[companies])
    # print('\n')
    if key in company_info[companies][1]:
        sydney_jobs += 1

print(sydney_jobs)        
```

```python
# Now we filter using keywords

python_jobs = results.find_all('h2', string='Python Developer')
print(python_jobs)
```

```python
# The python_jobs object turned up empty because we asked the code explicitly to find exact matches for 'Python Developer'
# We can use regex or a Lambda function to resolve this
# lambda functions are annonymous functions (Read about functional programming or lambda calculus)

python_jobs = results.find_all('h2', string=lambda text: 'python' in text.lower())

# The lambda function looks at the text of each <h2> element, 
# converts it to lowercase, and checks whether the substring 'python' is found anywhere in there

print(python_jobs)
```

```python
# Found one!
# We need to clean it up though
# If you look at it carefully thought, you can figure out that it is in Woodlands, WA and hte company is lancesoft

python_jobs = results.find_all('h2',string=lambda text: "python" in text.lower())

for p_job in python_jobs:
    link = p_job.find('a')['href']
    print(p_job.text.strip())
    print(f"Apply here: {link}\n")
```
