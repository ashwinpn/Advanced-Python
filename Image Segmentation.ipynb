```python
# This flag represents a conversion from BGR (Blue, Green, Red) to RGB
import matplotlib.pyplot as plt
import numpy as np
scooby = cv2.imread('1140-pet-anxiety-sad-dog.imgcache.rev3fb2e34d1eac84b2800e3174bb243597.jpg')
```

```python
plt.imshow(scooby)
plt.show()
```

```python
# By default OpenCV reads images in BGR format
# To fix this, we use cvtColor

scooby = cv2.cvtColor(scooby, cv2.COLOR_BGR2RGB)
plt.imshow(scooby)
plt.show()
```

```python
# The image is fixed now

from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib import colors
```

```python
r, g, b = cv2.split(scooby)
fig = plt.figure()
axis = fig.add_subplot(1, 1, 1, projection="3d")
```

```python
pixel_colors = scooby.reshape((np.shape(scooby)[0]*np.shape(scooby)[1], 3))
norm = colors.Normalize(vmin=-1.,vmax=1.)
norm.autoscale(pixel_colors)
pixel_colors = norm(pixel_colors).tolist()
```

```python
axis.scatter(r.flatten(), g.flatten(), b.flatten(), facecolors=pixel_colors, marker=".")
axis.set_xlabel("Red")
axis.set_ylabel("Green")
axis.set_zlabel("Blue")
plt.show()
```

```python
hsv_nemo = cv2.cvtColor(scooby, cv2.COLOR_RGB2HSV)
```

```python
h, s, v = cv2.split(hsv_nemo)
fig = plt.figure()
axis = fig.add_subplot(1, 1, 1, projection="3d")

axis.scatter(h.flatten(), s.flatten(), v.flatten(), facecolors=pixel_colors, marker=".")
axis.set_xlabel("Hue")
axis.set_ylabel("Saturation")
axis.set_zlabel("Value")
plt.show()
```
